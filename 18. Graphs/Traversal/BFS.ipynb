Traversal Idea (We will use the graph in pic):
- Initially we create an empty array called 'VISITED' filled with 0's of size n+1 where n = no. of nodes.
```py
VISITED = [0] * (n+1)
# n = 8 (0 to 7)
```
- When we get a node i.e, initially the root (1), we put the node in a Queue and mark that node index value as visited by turning the index value of VISITED to 1. For ex: if root's node value is 6 we turn index 6 of VISITED to 1.
```py
VISITED = [0, 1, 0, 0, 0, 0, 0, 0, 0]
Queue = [1]
```
- Next we pop the value of Queue and store it in a variable lets say 'e', also store it in RESULT array and check the left and right nodes connected to 'e' from adjacency list.
```py
Queue = []
e = 1
RESULT = [1]
# adjacency list : 1 --> [2, 3]
```
- We see if the left and right node of 'e' is visited or not by checking the index value in VISITED array. If not visited (value is 0) we put it in Queue (doesnot matter whether we put left first or right) and mark the index in VISITED to 1.
```py
# Before checking
VISITED = [0, 1, 0, 0, 0, 0, 0, 0, 0]

# After checking
VISITED = [0, 1, 1, 1, 0, 0, 0, 0, 0]
Queue = [2,3]
# adjacency list : 2 --> [1, 4, 5] ; 3 --> [1, 6, 7]
```
- Next we pop the value of Queue (2) and put it in RESULT array, and check what is e (2) connected to [1, 4, 5]. First 1 but in VISITED array its value is 1 (i.e, visited before) so we ignore it next 4 and 5 in VISITED is still 0 so we put it in our Queue. We do same for remaining node i.e.,node (3)
```py
# for 2
VISITED = [0, 1, 1, 1, 0, 0, 0, 0, 0]

e = 2

# after checking 3's adjacency list [1, 6, 7]
VISITED = [0, 1, 1, 1, 1, 1, 1, 1, 0]
Queue = [4, 5, 6, 7]
```
- Now we repeat i.e, pop value of Queue, store it in result, check adjacency list if left and right node visited ignore else put it in queue and mark it 1 in VISITED

````python
from collections import deque

def BFS(n, adj, starting_node):
    ans = []
    visited = [0] * (n + 1)
    queue = deque()
    queue.append(starting_node)
    visited[starting_node] = 1
    while queue != 0:
        e = queue.pop()
        ans.append(e)
        for node in adj[e]:
            if visited[node] == 0:
                queue.append(node)
                visited[node] == 1
    return ans
    

# n = 9
# adjacency_list = [
#     [],
#     [2, 8],
#     [1, 3, 4],
#     [2],
#     [2, 5],
#     [4, 6],
#     [5, 7],
#     [6, 8],
#     [1, 7, 9],
#     [8],
# ]
n = 8 
adjacency_list = [
[], 
[2, 3], 
[1, 4, 5],  
[1, 6, 7],  
[2],  
[2, 8],  
[3, 8],  
[3], 
[5, 6] 
]

print(bfs(n,adjacency_list,1))
````
